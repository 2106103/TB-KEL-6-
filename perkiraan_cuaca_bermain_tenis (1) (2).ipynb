```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("PERKIRAAN CUACA BERMAIN TENIS.csv", sep=';')
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
dataset['Humadity']= enc.fit_transform(dataset['Humadity'].values)
dataset['windy']= enc.fit_transform(dataset['windy'].values)
dataset['play']= enc.fit_transform(dataset['play'].values)
```

```python
dataset['Outlook']= enc.fit_transform(dataset['Outlook'].values)
dataset['Temperatur']= enc.fit_transform(dataset['Temperatur'].values)
```

```python
dataset
```

```python
atr_dataset = dataset.drop(columns = 'play')
atr_dataset
```

```python
cls_dataset = dataset['play']
cls_dataset
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(atr_dataset, cls_dataset, test_size=0.2, random_state=1)
tree_dataset = DecisionTreeClassifier(random_state=1)
tree_dataset.fit(xtrain, ytrain)
```

```python
y_pred = tree_dataset.predict(xtest)
cm = confusion_matrix(ytest, y_pred)
print("Confusion Matrix")
print(cm)
akurasi = classification_report(ytest, y_pred)
print("Tingkat Akurasi Algoritma")
print("Akurasi: ", akurasi)
akurasi = accuracy_score(ytest, y_pred)
print("Tingkat Akurasi: %d persen" %(akurasi*100))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree_dataset, out_file="Tree.play.dot", class_names=["A","B","C","D"],
                feature_names=atr_dataset.columns, impurity=False, filled=True)
```
